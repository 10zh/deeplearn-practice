```python
qW = nn.Linear(768, 64)
kW = nn.Linear(768, 64)
vW = nn.Linear(768, 64)
input = torch.randn(2, 768)
q = qW(input)
k = kW(input)
v = vW(input)
weight = q @ k.transpose(-2, -1)
print(weight)
print(v)
print(weight @ v)
```

```python
# 创建一个一维张量
a1 = torch.arange(0, 16)
print(a1)

# 使用-1让PyTorch自动计算维度
a2 = a1.view(-1, 16)
a3 = a1.view(-1, 8)
a4 = a1.view(-1, 4)
a5 = a1.view(-1, 2)
print(a2)
print(a3)
print(a4)
print(a5)
```
